## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
df['ConvertedComp'].plot(kind='box',figsize=(5,10))
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
a = df['ConvertedComp'].describe().loc['75%'] - df['ConvertedComp'].describe().loc['25%']
print('Inter Quartile Rangle: ',a)
```

Find out the upper and lower bounds.


```python
# your code goes here
print('Upper bound:',df['ConvertedComp'].max())
print('Lower bound:',df['ConvertedComp'].min())
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
condition2 = ((df['ConvertedComp'] > ( df['ConvertedComp'].quantile(0.75) + 1.5*a )) | (df['ConvertedComp'] < ( df['ConvertedComp'].quantile(0.25) - 1.5*a )))
print('there are {} outliners'.format(condition2.sum()))
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
import numpy as np
df2 = df
df2.shape
df2[condition2] = np.nan
df2.dropna(subset=['ConvertedComp'], axis=0, inplace=True)
```
